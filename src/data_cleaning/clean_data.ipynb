# Now, it's necessary to split columns like Red_Td
- pattern is: 'X of Y'
- thus, ' of ' is to be used for splitting 

```python
final_df[cols_to_split].head(1)
```

# As it can be seen above, only these columns contain 'of':

```python
cols_of = [t[0] for t in final_df[cols_to_split].head(1).items() if ' of ' in str(t[1])]
print(cols_of)
```

```python
def split_of(df, color):
    for col in cols_of:
        col_to_split = color + col
        print(col_to_split)
        success = df[col_to_split].str.split(' of ').str[0]
        attempt = df[col_to_split].str.split(' of ').str[1]
        df[col_to_split + '_Success'] = success
        df[col_to_split + '_Attempt'] = attempt
        # column is split, so we can safely drop it now
        df = df.drop(columns=[col_to_split]) 
    return df
```

```python
red  = split_of(red,  'Red_')
assert red.shape[1]  == len(tmp.columns)/2 + 2 * len(cols_of) - len(cols_of)
blue = split_of(blue, 'Blue_')
assert blue.shape[1] == len(tmp.columns)/2 + 2 * len(cols_of) - len(cols_of)
```

# As seen above, the only values to be replaced are:
- '--'
- '---'
- Note: there is some inconsistency on the UFC side. Sometimes '--'/'---' is interchangable with 0
- e.g.: both Red_Ctrl == '--' and Red_Ctrl == 0 mean that fighter in the red corner had no ground/clinch control whatsoever

```python
def get_clean_dict(df):
    to_clean = ['--', '---']
    d = dict()
    for val in to_clean:
        for col in df.columns:
            # if that value is present in the column we add it into the dict
            if len(df[df[col] == val]) != 0:
                # create a value-column_list pair if not present yet
                if val not in d.keys():
                    d[val] = []
                d[val].append(col)
                print(val + ' is present ' + f'{len(df[df[col] == val])} times in {col}')
    print('Returning dict:')
    print(d)
    return d
```

```python
red_clean_dict = get_clean_dict(red)
```

```python
blue_clean_dict = get_clean_dict(blue)
```

```python
def replace_values(df, d):
    for val, cols in d.items():
        for col in cols:
            df[col] = df[col].replace(val, '0')
    return df
```

```python
replace_values(red, red_clean_dict)
assert len(get_clean_dict(red)) == 0
```

```python
replace_values(blue, blue_clean_dict)
assert len(get_clean_dict(blue)) == 0
```
