```python
def name_match(f, r):
    for i in range(len(r)):
        if f[i] != r[i]:
            return False
    print(f, r)
    return True
```

```python
ssum = 0
final_df = pd.DataFrame()
for i in range(1, 25):
    # df_T - fight table dataframe
    df_T = pd.read_csv(f'../scraping/tables/page_{i}')
    # df_I - additional information about a fight dataframe
    df_I = pd.read_csv(f'../scraping/additional_info/page_{i}')
    # df_F - fighters info dataframe
    df_F = pd.read_csv(f'../scraping/fighters/page_{i}')
    
    # not counting in draws, only victories (and losses)
    mask = df_I['winner'] != 'Draw'
    df_I = df_I[mask].reset_index(drop=True);
    df_F = df_F[mask].reset_index(drop=True);
    tmp = pd.concat([df_I, df_F, df_T], axis=1)
    final_df = pd.concat([final_df, tmp], axis=0)
    ssum += tmp.shape[0]
    F = list(tmp['Fighter'].str.split(' '))
    R = list(tmp['Red_Name'].str.split(' '))
    for i in range(len(F)):
        assert name_match(F[i], R[i]) == True
ssum
```

```python
assert final_df.shape[0] == ssum
```
