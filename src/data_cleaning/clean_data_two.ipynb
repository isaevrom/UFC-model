# Let's work on referees now

- Since fights are sorted in chronological order, it is unlikely for referee to be the same in X years, as he could get, for example, fired by that moment
- Thus, I'll try to find the most frequent referee -75 and +75 fights away from the fight with a missing referee and impute the missing value with his name  

```python
df[df.referee.isna()].referee
```

```python
from random import randint

def impute_referee(df, row):
    most_freq_referees = df.iloc[row-75:row+75, :].referee.mode()
    n = len(most_freq_referees)
    # if there are more referees with same frequency, let randomness choose for us
    if n > 1:
        return most_freq_referees[randint(1,10) % n]
    return most_freq_referees[0]
```

```python
missing_referees_rows = df.index[df.referee.isna()].tolist()

for row in missing_referees_rows:
    df.loc[row, 'referee'] = impute_referee(df, row)
```

## Let's convert height into centimeters

```python
def height_to_cm(x):
    x = x.split(' ')
    feet = int(x[0].split('\'')[0])
    inches = int(x[1].split('"')[0])
    inches += feet * 12
    return inches * 2.54
```

```python
mask = ~df['Red_Height'].isna()
df.loc[mask, 'Red_Height'] = df[mask].Red_Height.apply(lambda x: height_to_cm(x))
```

```python
mask = ~df['Blue_Height'].isna()
df.loc[mask, 'Blue_Height'] = df[mask].Blue_Height.apply(lambda x: height_to_cm(x))
```

## Let's also convert reach into centimeters, while we are at it

```python
def reach_to_cm(x):
    inches = int(x.split('"')[0])
    return inches * 2.54
```

```python
mask = ~df['Red_Reach'].isna()
df.loc[mask, 'Red_Reach'] = df[mask].Red_Reach.apply(lambda x: reach_to_cm(x))
```

```python
df.Red_Reach = pd.to_numeric(df.Red_Reach, downcast='float')
```

```python
mask = ~df['Blue_Reach'].isna()
df.loc[mask, 'Blue_Reach'] = df[mask].Blue_Reach.apply(lambda x: reach_to_cm(x))
```

```python
df.Blue_Reach = pd.to_numeric(df.Blue_Reach, downcast='float')
```

## Finally imputing Red_Height

```python
df[df.Red_Height.isna()]
```

```python
mask = df['weight_division'] == 'Light Heavyweight Bout'
mean_lhb_weight = pd.concat([df[mask].Red_Height, df[mask].Blue_Height]).mean()
mean_lhb_weight
```

```python
df.loc[df.Red_Height.isna(), 'Red_Height'] = mean_lhb_weight
```
