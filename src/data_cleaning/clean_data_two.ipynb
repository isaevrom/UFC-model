# Let's work on referees now

- Since fights are sorted in chronological order, it is unlikely for referee to be the same in X years, as he could get, for example, fired by that moment
- Thus, I'll try to find the most frequent referee -75 and +75 fights away from the fight with a missing referee and impute the missing value with his name  

```python
df[df.referee.isna()].referee
```

```python
from random import randint

def impute_referee(df, row):
    most_freq_referees = df.iloc[row-75:row+75, :].referee.mode()
    n = len(most_freq_referees)
    # if there are more referees with same frequency, let randomness choose for us
    if n > 1:
        return most_freq_referees[randint(1,10) % n]
    return most_freq_referees[0]
```

```python
missing_referees_rows = df.index[df.referee.isna()].tolist()

for row in missing_referees_rows:
    df.loc[row, 'referee'] = impute_referee(df, row)
```

## Let's convert height into centimeters

```python
def height_to_cm(x):
    x = x.split(' ')
    feet = int(x[0].split('\'')[0])
    inches = int(x[1].split('"')[0])
    inches += feet * 12
    return inches * 2.54
```

```python
mask = ~df['Red_Height'].isna()
df.loc[mask, 'Red_Height'] = df[mask].Red_Height.apply(lambda x: height_to_cm(x)).astype(float)
```

```python
mask = ~df['Blue_Height'].isna()
df.loc[mask, 'Blue_Height'] = df[mask].Blue_Height.apply(lambda x: height_to_cm(x)).astype(float)
```

### Time length of the last round is already counted as 'time' feature. Hence we can get number of full-length rounds fought by substracting 1

```python
df['rounds'] = df['rounds'] - 1 
df['rounds']
```

```python
df[['rounds', 'time', 'rounds_format']]
```

```python
# function calculates number of seconds spent in full-length rounds
def full_rounds_time_spent(x):
    _sum = 0
    # n = number of full rounds
    n = int(x.split('[')[0])
    i = 0
    str_rounds = x.split('[')[1].strip(']')
    while n > 0:
        str_num = ''
        while str_rounds[i].isnumeric():
            str_num += str_rounds[i]
            i += 1
        if str_num != '':
            _sum += 60 * int(str_num)
            n -= 1
        i += 1
    return _sum
```

```python
full_rounds_time = df['rounds'].astype('str').str.cat(df['rounds_format'].astype('str')).apply(
    lambda x: full_rounds_time_spent(x)
)
full_rounds_time
```

```python
# function calculates maximum fight length in seconds  
def max_time(x):
    x = list(x)
    _sum = 0
    for mins_per_round in x:
        _sum += 60 * int(mins_per_round)
    return _sum
```

```python
max_time_col = df['rounds_format'].apply(lambda x: max_time(x))
max_time_col
```

```python
df['Fight_Length %'] = (df['time'] + full_rounds_time)/max_time_col * 100
df['Fight_Length %']
```
