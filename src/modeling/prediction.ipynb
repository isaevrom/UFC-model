```python
non_predictive_columns = [
    'method',
    'rounds', 'time', 'rounds_format',
    'Red_Record', 'Blue_Record',
    'Fight_Length %'
]

data = df.drop(columns=non_predictive_columns).copy()
```

# Encoding dependent variable
- Red corner fighter **won** the fight  => Winner value is 1
- Red corner fighter **lost** the fight => Winner value is 0

```python
y = data.apply(lambda x: x.Winner == x.Red_Name, axis=1).astype(int).copy()
y
```

# Preparing data for prediction
- X_A: 
    - applying One Hot Encoding on **all of the object features**
- X_B: 
    - applying OHE on **referee column only**, as it is probably the most important feature; dropping the rest
- X_C: 
    - applying OHE on **all of the object features**, but **dropping referee** column, as it is also the highest cardinality categorical feature, having more than 200 unique values
- X_D:
    - **dropping all** of the categorical features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = data.drop(columns=['Red_Name', 'Blue_Name', 'Winner']).copy()
X = X.reindex(sorted(X.columns), axis=1)
X.head(5).transpose();
```

```python
# X_A ... all
X_A = X
# X_B ... referee
X_B = X.drop(columns=['weight_division', 'Red_STANCE', 'Blue_STANCE'])
# X_C ... all - referee
X_C = X.drop(columns=['referee'])
# X_D ... droping all
categorical_cols = [col for col in data.drop(columns=['Red_Name', 'Blue_Name', 'Winner']) if df[col].dtype == 'O']
X_D = X.drop(columns=categorical_cols)
X_D = pd.concat([X_D], axis=1, sort=True)
```

```python
X_A = pd.get_dummies(X_A)
X_A;
# scaled version
scaler_mm = MinMaxScaler()
X_A_scaled = scaler_mm.fit_transform(X_A)
X_A_scaled;
```

```python
red_cols  = [col for col in data.columns if col.startswith('Red_')]
blue_cols = [col for col in data.columns if col.startswith('Blue_')]
```

```python
def get_winner(clf, red_name, blue_name):
    print(f'{red_name} vs. {blue_name}')
    
    red  = data[data.Red_Name.str.startswith(red_name)].head(1)[red_cols].drop(columns=['Red_Name', 'Red_STANCE'])
    if red.shape[0] == 0:
        red  = data[data.Blue_Name.str.startswith(red_name)].head(1)[blue_cols].drop(columns=['Blue_Name', 'Blue_STANCE'])
        red = red.rename(columns={i: 'Red_' + i[5:] for i in X_D.columns if i.startswith('Blue_') })
        
    blue = data[data.Blue_Name.str.startswith(blue_name)].head(1)[blue_cols].drop(columns=['Blue_Name', 'Blue_STANCE'])
    if blue.shape[0] == 0:
        blue  = data[data.Red_Name.str.startswith(blue_name)].head(1)[red_cols].drop(columns=['Red_Name', 'Red_STANCE'])
        blue = blue.rename(columns={i: 'Blue_' + i[4:] for i in X_D.columns if i.startswith('Red_') })
    
    red = red.reset_index(drop=True)
    blue = blue.reset_index(drop=True)
    
    fight = pd.concat([red, blue], axis=1, sort=True)
    fight = fight.reindex(sorted(fight.columns), axis=1)
    print(clf.predict(fight))
    return clf.predict(fight)
```
