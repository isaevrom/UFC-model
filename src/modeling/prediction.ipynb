```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data_cleaning/final data/data_2')
df.head(5).transpose()
```

```python
non_predictive_columns = [
    'method',
    'rounds', 'time', 'rounds_format',
    'Red_Record', 'Blue_Record',
    'Fight_Length %'
]

data = df.drop(columns=non_predictive_columns).copy()
```

```python
data.head(5).transpose()
```

# Encoding dependent variable
- Red corner fighter **won** the fight  => Winner value is 1
- Red corner fighter **lost** the fight => Winner value is 0

```python
y = data.apply(lambda x: x.Winner == x.Red_Name, axis=1).astype(int).copy()
y
```

# Preparing data for prediction
#### (Note: after playing around with it I ended up using scaled version of dataframe X_D)
- X_A: 
    - applying One Hot Encoding on **all of the object features**
- X_B: 
    - applying OHE on **referee column only**, as it is probably the most important feature; dropping the rest
- X_C: 
    - applying OHE on **all of the object features**, but **dropping referee** column, as it is also the highest cardinality categorical feature, having more than 200 unique values
- X_D:
    - **dropping all** of the categorical features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = data.drop(columns=['Red_Name', 'Blue_Name', 'Winner']).copy()
X = X.reindex(sorted(X.columns), axis=1)
X.head(5).transpose();
```

```python
# X_A ... all
X_A = X
# X_B ... referee
X_B = X.drop(columns=['weight_division', 'Red_STANCE', 'Blue_STANCE'])
# X_C ... all - referee
X_C = X.drop(columns=['referee'])
# X_D ... droping all
categorical_cols = [col for col in data.drop(columns=['Red_Name', 'Blue_Name', 'Winner']) if df[col].dtype == 'O']
X_D = X.drop(columns=categorical_cols)
X_D = pd.concat([X_D], axis=1, sort=True)
# also creating normalized version of X_D
scaler_mm = MinMaxScaler()
X_D_scaled = pd.DataFrame(scaler_mm.fit_transform(X_D),
                          index=X_D.index,
                          columns=X_D.columns)
```

```python
X_A = pd.get_dummies(X_A)
X_A;
```

```python
X_B = pd.get_dummies(X_B)
X_B;
```

```python
X_C = pd.get_dummies(X_C)
X_C;
```

```python
import seaborn as sns
import numpy as np

f, ax = plt.subplots(figsize=(10, 8))
corr = X.loc[:, ['Red_SLpM', 'Red_SApM', 'Red_TD Def.', 'Red_Winrate']].corr()
corr = X_D.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
plt.show()
```

# Trying different models:
- **Random Forest**
    - generally good and widely used model
- **XGBoost**
    - this model is shown to be successful particularly for predictions of outcomes of matches, as its good at finding complex relationships in data
    - many Kaggle competitions winners owe their victory to XGBoost

```python
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb

from sklearn.model_selection import train_test_split

from sklearn.model_selection import RandomizedSearchCV, GridSearchCV

from sklearn import metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
X_D.head(5).transpose()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_D_scaled,
                                                    y,
                                                    test_size=0.25, 
                                                    random_state=123,
                                                    stratify=y)
```

## Random Forest Classifier

* Tuning model by finding best hyperparameters

```python
rfc  = RandomForestClassifier()
params = {
 'bootstrap': [True, False],
 'max_depth': [5, 10, 15],
 'max_features': ['sqrt'],
 'min_samples_leaf': [1, 2, 4, 8],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [50, 75, 150, 250]
}
rfc_random = RandomizedSearchCV(estimator=rfc, 
                                param_distributions=params, 
                                n_iter=25, 
                                cv=5, 
                                scoring='roc_auc',
                                n_jobs=4,
                                verbose=1, 
                                random_state=42)
rfc_random.fit(X_train, y_train)
```

* Best hyperparameters for **Random Forest Classifier**:

```python
best_params = rfc_random.best_params_
best_params
```

```python
best_rfc = RandomForestClassifier(**best_params)
best_rfc.fit(X_train, y_train)
```

```python
y_pred = best_rfc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred, labels=sorted(best_rfc.classes_, reverse=True)))
```

```python
# saving accuracy of RF
rf_acc = accuracy_score(y_test, y_pred)
# saving average f1-score of RF
rf_score_dict = classification_report(y_test, y_pred, 
                                      labels=sorted(best_rfc.classes_, reverse=True), 
                                      output_dict=True)
rf_f1 = rf_score_dict['macro avg']['f1-score']

print(f'accuracy: {rf_acc}')
print(f'f1-score: {rf_f1}')
```

## XGBoost

* Tuning model by finding best hyperparameters

```python
xgbc = xgb.XGBClassifier(learning_rate=0.02, objective='binary:logistic')
params = {
        'min_child_weight': [1, 5, 10, 15],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5],
        'n_estimators': [150, 250, 350]
}
random_search = RandomizedSearchCV(xgbc,
                                   param_distributions=params, 
                                   n_iter=5, 
                                   cv=5,
                                   scoring='roc_auc', 
                                   n_jobs=4, 
                                   verbose=1, 
                                   random_state=1001)
random_search.fit(X_train, y_train)
```

* Best hyperparameters for **XGBoost**:

```python
best_params_xgbc = random_search.best_params_
best_params_xgbc
```

```python
best_xgbc = xgb.XGBClassifier(learning_rate=0.02, 
                              objective='binary:logistic',
                              silent=True,
                              **best_params_xgbc)
best_xgbc.fit(X_train, y_train)
```

```python
y_pred = best_xgbc.predict(X_test)
xgb_acc = accuracy_score(y_test, y_pred)
xgb_f1  = f1_score(y_test, y_pred)
```

```python
print(f'accuracy: {xgb_acc}')
print(f'f1-score: {xgb_f1}')
print(classification_report(y_test, y_pred, labels=sorted(best_xgbc.classes_, reverse=True)))
```

```python
# saving accuracy of XGB
xgb_acc = accuracy_score(y_test, y_pred)
# saving average f1-score of XGB
xgb_score_dict = classification_report(y_test, y_pred,
                                       labels=sorted(best_xgbc.classes_, reverse=True),
                                       output_dict=True)
xgb_f1 = xgb_score_dict['macro avg']['f1-score']

print(f'accuracy: {xgb_acc}')
print(f'f1-score: {xgb_f1}')
```

## Comparing the two

* As we can see below, both of the models are pretty much equally good when it comes to the **False Positives** and **True Positives** ratio. With slightly better score for XGBoost

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

classifiers = [best_rfc, best_xgbc]

model_names = {
    best_rfc: 'RandomForest',
    best_xgbc:'XGBoost',
}

plt.figure(figsize=(5,5), dpi=150)

for clf in classifiers:
    y_pred = clf.predict_proba(X_test)[:,1] 
    
    fpr, tpr, threshhold = roc_curve(y_test, y_pred)
    _auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr, linestyle='-', 
             label=f'{model_names[clf]}: AUC = {_auc:.3f}')
    
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.legend()
plt.grid()
plt.show()
```

* Let's take a look at **f1-score** and **accuracy**:
    * As shown below, XGBoost has slightly better both accuracy & f1-score 

```python
x_axis = np.arange(2)
labels = ['Random Forest','XGBoost']

accs = [rf_acc, xgb_acc]
f1s  = [rf_f1, xgb_f1]

plt.figure(figsize = (12, 7.5)) 

# bars for f1-score
bars_f1 = plt.bar(x_axis - 0.2, 
        f1s,  
        width=0.4,  
        label = 'f1-score (avg)')

for bar in bars_f1:
    score = bar.get_height()
    plt.text(bar.get_x() + 0.15, score + .005, np.round(score, 4))

# bars for accuracy
bars_acc = plt.bar(x_axis + 0.2, 
        accs, 
        width=0.4,
        label = 'accuracy')

for bar in bars_acc:
    score = bar.get_height()
    plt.text(bar.get_x() + 0.15, score + .005, np.round(score, 4))

plt.xticks(x_axis, labels)
plt.xlabel("Model")
plt.ylabel("Score of metric")
plt.title("Impact of model choice on classification outcome")
plt.legend()
plt.show()
```

# Bonus: UFC-280 predictions
* Models have not seen UFC-280 data yet, as that data was not present at time of scraping
* Let's see how well they predict outcomes of UFC-280 fights

```python
red_cols  = [col for col in data.columns if col.startswith('Red_')]
blue_cols = [col for col in data.columns if col.startswith('Blue_')]
```

```python
def get_winner(clf, red_name, blue_name):
    print(f'{red_name} vs. {blue_name}')
    
    red  = data[data.Red_Name.str.startswith(red_name)].head(1)[red_cols].drop(columns=['Red_Name', 'Red_STANCE'])
    if red.shape[0] == 0:
        red  = data[data.Blue_Name.str.startswith(red_name)].head(1)[blue_cols].drop(columns=['Blue_Name', 'Blue_STANCE'])
        red = red.rename(columns={i: 'Red_' + i[5:] for i in X_D_scaled.columns if i.startswith('Blue_') })
        
    blue = data[data.Blue_Name.str.startswith(blue_name)].head(1)[blue_cols].drop(columns=['Blue_Name', 'Blue_STANCE'])
    if blue.shape[0] == 0:
        blue  = data[data.Red_Name.str.startswith(blue_name)].head(1)[red_cols].drop(columns=['Red_Name', 'Red_STANCE'])
        blue = blue.rename(columns={i: 'Blue_' + i[4:] for i in X_D_scaled.columns if i.startswith('Red_') })
    
    red = red.reset_index(drop=True)
    blue = blue.reset_index(drop=True)
    
    fight = pd.concat([red, blue], axis=1, sort=True)
    fight = fight.reindex(sorted(fight.columns), axis=1)
    print(clf.predict(pd.DataFrame(scaler_mm.transform(fight),
                                   columns=X_D_scaled.columns)))
    return clf.predict(pd.DataFrame(scaler_mm.transform(fight),
                                   columns=X_D_scaled.columns))
```

```python
def event_prediction(clf, event):
    list_pred = []
    # getting prediction for each fight
    for fight in ufc280:
        prediction = get_winner(clf=clf, 
                                red_name=fight[0], 
                                blue_name=fight[1])
        # using idx [0] just to transform one-element array to number
        list_pred.append(prediction[0])
    number_of_fights = len(event)
    # real_outcomes is just an array of ones
    real_outcomes = [1 for i in range(number_of_fights)]
    # indicates number of fights which outcome prediction is correct
    correct = 0
    for i in range(len(real_outcomes)):
        if list_pred[i] == real_outcomes[i]:
            correct += 1
    ratio = correct/float(number_of_fights) * 100
    print(f'{ratio}% of fight outcomes predicted correctly')
    return ratio 
```

```python
ufc280 = [
    ['Islam Makhachev', 'Charles Oliveira'],
    ['Aljamain Sterling', 'TJ Dillashaw'],
    ['Sean O\'Malley', 'Petr Yan'],
    ['Beneil Dariush', 'Mateusz Gamrot'],
    ['Manon Fiorot', 'Katlyn Chookagian'],
    ['Belal Muhammad', 'Sean Brady'],
    ['Caio Borralho', 'Makhmud Muradov'],
    ['Nikita Krylov', 'Volkan Oezdemir'],
    ['Armen Petrosyan', 'AJ Dobson'],
    ['Abubakar Nurmagomedov', 'Gadzhi Omargadzhiev'],
    ['Muhammad Mokaev', 'Malcolm Gordon'],
    ['Karol Rosa', 'Lina Lansberg']
]
```

```python
def plot_pie(ratio, clf_name):
    correct = ratio
    incorrect = 100 - ratio
    sizes = [correct, incorrect]
    
    explode = (0, 0.125)  # only "explode" the 2nd slice (i.e. 'Hogs')
    
    fig, ax = plt.subplots()
    ax.pie(sizes,
           explode=explode,
           shadow=True,
           startangle=90,
           autopct='%1.2f%%')
    plt.title(f'UFC-280 predictions ratio ({clf_name})')
    plt.legend(['Correctly predicted', 'Incorrectly predicted'])
    plt.show()
```

## Random Forest

```python
rf_ratio = event_prediction(best_rfc, ufc280)
plot_pie(rf_ratio, clf_name='Random Forest')
```

## XGBoost

```python
xgb_ratio = event_prediction(best_xgbc, ufc280)
plot_pie(xgb_ratio, clf_name='XGBoost')
```
